```python
image1=cv.imread("C:/Users/DELL/Downloads/Mit.jpg")[:,:,::-1]#<- this is for image in original form rgb
b,g,r=cv.split(image1) # this will split image in 3 two dimentional arrays
```

```python
plt.figure(figsize=(30,5))
plt.subplot(1,4,1);plt.imshow(image1,cmap="gray")
plt.subplot(1,4,2);plt.imshow(b,cmap="gray") # you can see some features are prominant in certain color
plt.subplot(1,4,3);plt.imshow(g,cmap="gray")
plt.subplot(1,4,4);plt.imshow(r,cmap="gray")
```

```python
# converting color of image from bgr to rgb(you can change it into gray or hsv as well)
image3=cv.cvtColor(image1,cv.COLOR_BGR2RGB)
plt.imshow(image3,cmap="gray") # you can clearly see red part is now blue
```

```python
image4=cv.cvtColor(image1,cv.COLOR_BGR2HSV)
plt.imshow(image4,cmap="gray") # this shows image in Hue, saturation and value format
```

```python
#h is heuristics, s is saturation and v is value
h,s,v=cv.split(image4)
plt.figure(figsize=(18,5))
plt.subplot(1,3,1);plt.imshow(h,cmap="gray");plt.title("h")
plt.subplot(1,3,2);plt.imshow(s,cmap="gray");plt.title("s") 
plt.subplot(1,3,3);plt.imshow(v,cmap="gray");plt.title("v") 
```

```python
h_new=h+10 # here you can make changes to individual values (you can do it same for s and v as well as for b, g, r)
image5=cv.merge((h_new,s,v))
image5=cv.cvtColor(image5,cv.COLOR_HSV2RGB)
plt.imshow(image5)
```

```python
# to save image (you can also provide path so to decide where to save image.)
cv.imwrite("MIT.jpg", image5)
```

## Image Annotation

```python
# Draing circle
resized_copy1=resized_image.copy()
cv.circle(resized_copy1,(700,1100),200,(255,0,0),5,cv.LINE_AA)
plt.imshow(resized_copy1)
```

```python
# Rectangle
resized_copy2=resized_image.copy()
cv.rectangle(resized_copy2,(16,1000),(1500,1400),(255,0,255),16,cv.LINE_AA)
plt.imshow(resized_copy2)
```

```python
# Writing text on image
resized_copy3=resized_image.copy()
cv.putText(resized_copy3,"MIT-WPU",(25,900),cv.FONT_HERSHEY_SIMPLEX,5,(255,255,255),5,cv.LINE_AA)
plt.imshow(resized_copy3)
```

## Image enhancement

```python
import numpy as np
resizedcopy4=resized_image.copy()
matrix=np.ones(resizedcopy4.shape,dtype='uint8')*50
```

```python
# Adding brightness
imagelight=cv.add(resizedcopy4,matrix)
plt.imshow(imagelight,cmap="gray")
```

```python
# Reducing brightness
matrix=np.ones(resizedcopy4.shape,dtype='uint8')*100
imagedark=cv.subtract(resizedcopy4,matrix)
plt.imshow(imagedark)
```

```python
# Adding contrast to image
matrix2=np.ones(resizedcopy4.shape)*1.2
img_rgb_dark=np.uint8(cv.multiply(np.float64(resizedcopy4),matrix2))
plt.imshow(img_rgb_dark)
```

```python
img_rgb_dark1=np.uint8(np.clip(cv.multiply(np.float64(resizedcopy4),matrix2),0,255))
plt.imshow(img_rgb_dark1)
```

```python
#image thresholding or masking used for processing certain part of image while keeping remaining part intact

retval,imgthreshold=cv.threshold(resizedcopy4,150,255,cv.THRESH_BINARY)
plt.subplot(1,2,1);plt.imshow(resizedcopy4)
plt.subplot(1,2,2);plt.imshow(imgthreshold,cmap="gray")
```

### In general pixel below certain value will be set to zero(black) and above certain value will be set to 255 (white)
### works better on black and white image (grayscale image)

```python
retval,imgthreshold=cv.threshold(resizedcopy4,100,255,cv.THRESH_BINARY)
plt.subplot(1,2,1);plt.imshow(resizedcopy4)
plt.subplot(1,2,2);plt.imshow(imgthreshold,cmap="gray")
```

```python
# Adaptive threshold
bw = cv.cvtColor(resizedcopy4, cv.COLOR_BGR2GRAY)

dst=cv.adaptiveThreshold(bw,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,11,7)
plt.subplot(1,2,1);plt.imshow(bw,cmap="gray")
plt.subplot(1,2,2);plt.imshow(dst,cmap="gray")
```

## Bitwise operation on image

```python

img7 = cv.imread("C:/Users/DELL/Downloads/download4.png",cv.IMREAD_GRAYSCALE)
width=1320
height=879
dim=(width,height)
Resized_img7=cv.resize(img7,dim,interpolation=cv.INTER_AREA)
plt.imshow(Resized_img7,cmap="gray")
```

```python
img6 = cv.imread("C:/Users/DELL/Downloads/download3.jpg",cv.IMREAD_GRAYSCALE)
plt.imshow(img6,cmap="gray")
img6.shape
```

```python
# Bitwise And

bit_and=cv.bitwise_and(Resized_img7,img6,mask=None)
plt.figure(figsize=(15,15))
plt.subplot(1,3,1);plt.imshow(Resized_img7,cmap="gray")
plt.subplot(1,3,2);plt.imshow(img6,cmap="gray")
plt.subplot(1,3,3);plt.imshow(bit_and,cmap="gray")
```

```python
# Bitwise Or

bit_or=cv.bitwise_or(Resized_img7,img6,mask=None)
plt.figure(figsize=(15,15))
plt.subplot(1,3,1);plt.imshow(Resized_img7,cmap="gray")
plt.subplot(1,3,2);plt.imshow(img6,cmap="gray")
plt.subplot(1,3,3);plt.imshow(bit_or,cmap="gray")
```

```python
# Bitwise Xor (combination of not and or)

bit_xor=cv.bitwise_xor(Resized_img7,img6,mask=None)
plt.figure(figsize=(15,15))
plt.subplot(1,3,1);plt.imshow(Resized_img7,cmap="gray")
plt.subplot(1,3,2);plt.imshow(img6,cmap="gray")
plt.subplot(1,3,3);plt.imshow(bit_xor,cmap="gray")
```

```python
#Not operator

bit_not=cv.bitwise_not(Resized_img7,mask=None)
plt.figure(figsize=(15,15))
plt.subplot(1,3,1);plt.imshow(Resized_img7,cmap="gray")
plt.subplot(1,3,3);plt.imshow(bit_not,cmap="gray")
```
